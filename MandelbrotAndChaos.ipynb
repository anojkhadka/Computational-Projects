### Mandelbrot Set

Mandelbrot set is a set of complex numbers that obey certain mathematical rule. First studied by Robert Brooks and Peter Matelski and visualized by Benoit Mandelbrot while woking in IBM, it represents the complexity and has many applications in the study of chaos theory, number theory, fractal geometry, etc. Mathematically, it is defined by a simple function:

$$x_{n} = x_{0}^2 + c  $$

where 'c' is a set of complex numbers and '$x_{0}$' is 0. This set of such complex numbers for which the functional value defined above does not diverge even at infinity, comprises a Mandelbrot set. Starting with $x_{0} = 0$ and any complex number 'c` we keep summing up the numbers iteratively as:

$$ x_{1} = 0 + c $$
$$ x_{2} = x_{1}^2 + c $$
$$ x_{3} = x_{2}^2 + c $$
$$..........$$
$$..........$$
$$..........$$

$$ x_{n} = x_{n-1}^2 + c $$

Even when we take this sum to infinity if the sum does not diverge the complex number 'c` is said to lie in a Mandelbrot set. Here the $x_{0}$ is called a seed and the summation orbit of the seed. So, Mandelbrot set is a set of these complex numbers for which the orbit of the seed 0 does not diverge to infinity. 

When we make a plot of Mandelbrot set, it gives rise to form a beautiful looking design pattern as shown below. Generally, we use complex nubers $z = x + iy$ within x and y values ranging in between (-2,2). It has been derived mathematically that this series will converge below a radius of 2 when seed is 0 for given set of complex numbers in above range of x and y values. So, we use this radius as a braking/limiting condition for our iteration loop. 

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot_set(lower_limit, upper_limit, total_number, max_iteration):
    
    """ This function will first create a 2-D array of zeros. We then create complex numbers within a complex plane with x and y 
    values ranging between (-1.5, 1.5). These complex numbers are each tested to see if iterative sum exceeds the radius of 
    convergence value 2 when seed of 0 is taken. If value diverge we change that complex number in the complex plane to 1.
    Finally a plot of these 2-D matrix set is created to visualize the mandelbrot set. One can tweak the seed value and range of
    complex numbers taken just to see how the graph changes."""
    
    blank_set = np.zeros((total_number,total_number))
    number_sample = np.linspace(lower_limit, upper_limit, total_number)
    for n1 in range(total_number):
        x = number_sample[n1]
        for n2 in range(total_number):
            y = number_sample[n2]
            c = complex(x,y)                              # Complex number to test
            z = 0                                         # Seed number
            for k in range(max_iteration):
                z = z**2 + c
                if abs(z) > 2:
                    blank_set[n1,n2] = 1
                    break
    return blank_set


def mandelbrot_graph(mandel_set):
    plt.figure(figsize=(12,12))
    plt.imshow(mandel_set, extent=(-2.0, 2.0, -2.0, 2.0))

def main():
    result1 = mandelbrot_set(-1.5, 1.5,1024,300)
    result2 = mandelbrot_graph(result1)

if __name__ == "__main__":
    main()
```
