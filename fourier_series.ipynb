```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-12,12,2)
y = x
x
```

```python
fig, fourier_plot = plt.subplots(figsize = (15,10))
for n in np.arange(5):
    m = 2*n + 1
    if (x[m] + x[m + 2])/4 == 0:
        x1 = np.linspace(x[m], x[m + 2],10)
        y1 = x1
        fourier_plot.plot(x1,y1)
        plt.hold(True)
```
